```python
import csv
import numpy as np
# l = ('E:\UF 21\Fall 22\Comp Tools\Problem Sets\CO-OPS__8729108__wl.csv')

with open('CO-OPS__8729108__wl.csv', 'r') as csv_file:
    csv_reader = csv.reader(csv_file)
    Water_Level = []
    Date = []
    Time = []
    for line in csv_reader:
        Water_Level.append(line[2])
        Date.append(line[0])
        Time.append(line[1])

# Water_Level[1:]
print('Highest water level: ', max(Water_Level[1:]))
A = np.argmax(Water_Level[1:])
B = np.argmin(Water_Level[1:])
print('Date and time for highest water level: ', Date[A+1], Time[A+1])
print('Lowest water level: ', min(Water_Level[1:]))
print('Date and time for lowest water level: ', Date[B+1], Time[B+1])
print('Average water level: ', np.mean(np.asarray(Water_Level[1:]), dtype='float'))
```
